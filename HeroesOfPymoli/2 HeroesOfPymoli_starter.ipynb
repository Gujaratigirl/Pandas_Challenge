## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Find total count of SN - counted only once per occurance- will use this DF to do caclulations
sn_count_df = purchase_data_df.groupby('SN').head(1)
sn_count = len(sn_count_df["SN"])
sn_count

#Make df by Gender
male_gender_df =sn_count_df.loc[purchase_data_df["Gender"] == "Male",:]
female_gender_df =sn_count_df.loc[purchase_data_df["Gender"] == "Female",:]
other_gender_df =sn_count_df.loc[purchase_data_df["Gender"] == "Other / Non-Disclosed",:]

#Count how many in each
male_count = len(male_gender_df)
female_count = len(female_gender_df)
other_count = len(other_gender_df)
#Percent of each
male_percent = round(male_count/sn_count*100,2)
female_percent = round(female_count/sn_count*100,2)
other_percent = round(other_count/sn_count*100,2)
#Make DF
gender_demo_df = pd.DataFrame({"Total Count":["",male_count,female_count, other_count],
                                "Percentage of Players": ["",str(male_percent)+"%",str(female_percent)+"%",str(other_percent)+"%"]},
                             ["Gender","Male", "Female", "Other / Non-Disclosed"])

gender_demo_df = gender_demo_df.style.set_properties(**{'text-align': 'left'})                       
gender_demo_df
```
